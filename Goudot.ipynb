# Test Matplotlib

Mission : Pareto ‘production’ pour ‘blé’

```python
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
prod = dfc[dfc['Code Élément']=='5511'] # production
prod = prod[prod['Code Produit']=='2511'] # Blé
prod = prod[prod['Symbole']=='S'] # Standard

prod=prod.sort_values(by=['Valeur'], ascending=False)
prod = prod[['Pays', 'Valeur']][:20]
prod
```

```python
plt.style.use('_mpl-gallery')
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]

# plot
fig, ax = plt.subplots()

prod=prod.sort_values(by=['Valeur'], ascending=True)
#ax.title('test')
ax.barh(prod.Pays, prod.Valeur) # , width=1, edgecolor="white", linewidth=0.7

#ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
