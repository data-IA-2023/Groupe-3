```python
# annee probem column



import pandas as pd


#These lines of code import the pandas library and load two CSV files ('FAOSTAT_2013_animal.csv' and 'FAOSTAT_2013_vegetal.csv') 
#into DataFrames named animal_data and vegetal_data.
# Load data
animal_data = pd.read_csv('fao_2013/FAOSTAT_2013_animal.csv')     
vegetal_data = pd.read_csv('fao_2013/FAOSTAT_2013_vegetal.csv')

# Merge data
#This line merges the two DataFrames (animal_data and vegetal_data) on the specified columns 
#('Code Pays', 'Code Produit', 'Année', 'Unité') using an outer join. The result is stored in a new DataFrame named merged_data.
merged_data = pd.merge(animal_data, vegetal_data, how='outer', on=['Code Pays', 'Code Produit', 'Année', 'Unité'])

# For the 20 most calorically-dense foods
#These lines calculate the energy density for each row by dividing the 'Valeur_x' column 
#(presumably representing energy) by the 'Valeur_y' column (presumably representing weight or quantity). #
#Then, it selects the top 20 rows with the highest energy density and stores them in the DataFrame top_caloric_foods.
merged_data['Energy Density'] = merged_data['Valeur_x'] / merged_data['Valeur_y']
top_caloric_foods = merged_data.sort_values(by='Energy Density', ascending=False).head(20)

# For the 20 foods richest in protein
#These lines select the top 20 rows with the highest values in the 'Valeur_x' column 
#(presumably representing protein content) and stores them in the DataFrame top_protein_foods.top_protein_foods
top_protein_foods = merged_data.sort_values(by='Valeur_x', ascending=False).head(20)


#These lines print the results. The first print statement displays the top 20 most calorically-dense foods with columns 
#'Pays_x', 'Produit_x', and 'Energy Density'. The second print statement displays the top 20 foods richest in protein 
#with columns 'Pays_x', 'Produit_x', and 'Valeur_x'.
# Display the results
print("Top 20 Most Calorically-Dense Foods:")
print(top_caloric_foods[['Pays_x', 'Produit_x', 'Energy Density']])

print("\nTop 20 Foods Richest in Protein:")
print(top_protein_foods[['Pays_x', 'Produit_x', 'Valeur_x']])




```

Établissez la liste des produits (ainsi que leur code) considéré comme des céréales selon la FAO. En ne prenant en compte que les céréales destinées à l'alimentation (humaine et animale), quelle proportion (en termes de poids) est destinée à l'alimentation animale ?

```python
#The code begins by importing the pandas library and giving it the alias pd.
#It then loads the dataset from the 'fao_2013/FAOSTAT_2013_cereal.csv' file using the pd.read_csv function and stores 
#it in a DataFrame named cereal_df.
#The column names of the dataset are then printed to the console.

import pandas as pd

# Load the cereal dataset
cereal_df = pd.read_csv('fao_2013/FAOSTAT_2013_cereal.csv')

# Display the columns in the dataset
print("Cereal Dataset Columns:")
print(cereal_df.columns)


#The code extracts the columns 'Produit' (product) and 'Code Produit' (product code) from the cereal_df DataFrame.
#It then removes duplicate rows to get a list of unique cereal products and their corresponding product codes.
#The result is printed to the console.
# Extract unique cereal products and their codes
cereal_products = cereal_df[['Produit', 'Code Produit']].drop_duplicates()



# Display the list of products considered as cereals
print("\nList of Cereal Products and Their Codes:")
print(cereal_products)


#The code filters the DataFrame to include only rows where the value in the 'Élément' column is 'Nourriture' (food).
#This creates a new DataFrame named cereals_for_food containing cereals intended for food.
# Filter cereals intended for food (human and animal)
cereals_for_food = cereal_df[cereal_df['Élément'] == 'Nourriture']

#The code calculates the total weight of cereals intended for food by summing the 'Valeur' (value) column in the cereals_for_food DataFrame.
#It also calculates the total weight of cereals intended for animal feed by subtracting the total weight for food from the overall total weight.
# Calculate the total weight of cereals intended for food and animal feed
total_weight_for_food = cereals_for_food['Valeur'].sum()
total_weight_for_animal_feed = cereal_df['Valeur'].sum() - total_weight_for_food


#The code calculates the proportion of total weight intended for animal feed as a percentage of the overall total weight.
# Calculate the proportion intended for animal feed
proportion_for_animal_feed = (total_weight_for_animal_feed / cereal_df['Valeur'].sum()) * 100


#The code prints the results, including the total weight of cereals intended for food, total weight for animal feed,
#and the proportion intended for animal feed as a percentage.
# Display the results
print("\nTotal Weight of Cereals Intended for Food:", total_weight_for_food, "kg")
print("Total Weight of Cereals Intended for Animal Feed:", total_weight_for_animal_feed, "kg")
print("Proportion Intended for Animal Feed:", proportion_for_animal_feed, "%")



```

#### Question 14 : Thaïlande
En Thaïlande, quelle proportion de manioc est exportée ? Quelle est la proportion de personnes en sous-nutrition?

```python
import pandas as pd

# Load the vegetal dataset
vegetal_data = pd.read_csv('fao_2013/FAOSTAT_2013_vegetal.csv', encoding='latin1')

# Filter data for Thailand and Cassava
thailand_cassava = vegetal_data[(vegetal_data['Pays'] == 'Thailand') & (vegetal_data['Item'] == 'Cassava')]

# Calculate the proportion of cassava exported in Thailand
# Proportion of cassava exported = (Export Quantity / Production Quantity) * 100
proportion_exported = (thailand_cassava['Export Quantity'] / thailand_cassava['Production']).iloc[0] * 100

# Print the result
print(f"The proportion of Cassava exported in Thailand: {proportion_exported:.2f}%")

# Note: The 'Undernourishment' column indicates the proportion of people in undernourishment.
# You can use this information directly.
undernourishment_thailand = vegetal_data[vegetal_data['Pays'] == 'Thailand']['Undernourishment'].iloc[0]
print(f"The proportion of people in undernourishment in Thailand: {undernourishment_thailand:.2f}%")
```
